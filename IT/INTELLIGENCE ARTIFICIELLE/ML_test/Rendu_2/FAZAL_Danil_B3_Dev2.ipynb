#### Exercice 2.3: Séparez les données en 2 échantillons train-test

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_curve, auc, precision_recall_curve
```

```python
data.dropna(inplace=True)

y = data['TARGET_5Yrs']
X = data.drop(labels=['TARGET_5Yrs'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)
```

#### Question 2.3: Pourquoi cette étape est-elle indispensable?

Elle est indispensable car elle permettra de vérifier la précision du machine learning.

#### Exercice 2.4: Entrainez sur ces données un arbre de décision. Calculer une prédiction.

```python
# calcul de prédiction
clf = DecisionTreeClassifier()

clf = clf.fit(X_train, y_train)
estimate = clf.predict(X_test)

print(abs(y_test - estimate).mean())
print((y_test - estimate).std())
```

#### Exercice 2.5: Evaluez la qualité du modèle. 

```python
accuracy_score(y_test, estimate, normalize=True, sample_weight=None)
```

#### Question 2.4: Quelle métrique utilisez vous? Pourquoi? 

J'utilise ici accuracy_score() pour décrire les individus positifs et négatifs symétriquement.

#### Question 2.5: Le modèle construit est-il de qualité satisfaisante? Expliquez.

On voit que la qualité du modèle construit est un peu insatisfaisante, on prédit bien les positifs et moins bien les négatifs. 

#### Exercice 2.6: Produisez une visualisation graphique de l'arbre construit

```python
# affichage de l'arbre
import matplotlib.pyplot as plt

plt.figure(figsize=(17,11))
plot_tree(clf, feature_names = list(X_train.columns), fontsize=10)
plt.show()
```

#### Question 2.6: Donnez un exemple de lecture d'un tel arbre.

On va parcourir les branches selon nos données. Par exemple si l'on a GP = 52.5 et FG% = 26, on va se diriger deux fois vers la branche de gauche car nos deux valeurs seront inférieurs à celles indiquées sur les feuilles de l'arbre.

#### Question 2.7: Comment peut - on améliorer la qualité du modèle? Donnez au moins 3 méthodes:

On peut faire une validation croisée, ajouter des hyperparamètres.

#### Exercice 2.7: Essayer d'appliquer une de ces méthodes

```python
# nouveau calcul en appliquant la méthode
clf2 = DecisionTreeClassifier(criterion= 'entropy', max_depth=5, min_samples_split=0.1)

clf2 = clf2.fit(X_train, y_train)
estimate2 = clf2.predict(X_test)
true = y_test

print(abs(estimate2 - true).mean())
print((estimate2 - true).std())

# affiche du nouvel arbre
plt.figure(figsize=(17,11))
plot_tree(clf2, feature_names = list(X_train.columns), fontsize=10)
plt.show()
```

#### Question 2.8: Concluez

```python
accuracy_score(y_test, estimate2, normalize=True, sample_weight=None)
```

En remesurant maintenant la qualité du modèle, on voir que le taux mesuré est passé d'environ 63.7% à 68.2%.
