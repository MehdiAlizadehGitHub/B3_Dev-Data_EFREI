#### Question 1.3: Quelles sont les 2 principales familles "d'apprentissage"? Donner une rapide définition de chacune

Réponse :

L'apprentissage supervisé : dans l'apprentissage supervisé, les algorithmes de machine learning sont entraînés sur une base de données d'exemples qui comprend des entrées (également appelées caractéristiques ou variables) et des sorties (également appelées étiquettes ou cibles). L'objectif de l'apprentissage supervisé est de prédire les sorties pour de nouvelles entrées en utilisant les exemples de la base de données. Par exemple, un modèle de machine learning supervisé peut être entraîné sur des données comprenant des caractéristiques d'une voiture (comme la taille du moteur, le poids, etc.) et l'étiquette indiquant si la voiture est considérée comme sportive ou non. L'objectif du modèle est de prédire si une nouvelle voiture est sportive ou non en fonction de ses caractéristiques.

L'apprentissage non supervisé : dans l'apprentissage non supervisé, les algorithmes de machine learning sont entraînés sur une base de données qui ne comprend pas d'étiquettes ou de cibles prédéfinies. L'objectif de l'apprentissage non supervisé est de trouver des structures ou des relations dans les données de manière automatique. Par exemple, un modèle d'apprentissage non supervisé peut être entraîné sur une base de données de transactions bancaires et utilisé pour détecter des groupes de transactions similaires (comme des dépenses de nourriture ou de loisirs).

#### Question 1.4: Dans laquelle de ces familles appartient la "classification"? Le "clustering"?

Réponse : 

La classification appartient à la famille de l'apprentissage supervisé, tandis que le clustering appartient à la famille de l'apprentissage non supervisé.

La classification : la classification est une tâche de machine learning qui consiste à prédire l'appartenance d'une observation à une classe prédéfinie en fonction de ses caractéristiques. Par exemple, un modèle de classification peut être entraîné pour prédire si une personne est atteinte d'une maladie ou non en fonction de ses symptômes et de ses résultats de laboratoire.

Le clustering : le clustering est une tâche de machine learning qui consiste à regrouper des observations similaires en "clusters" (groupes). L'objectif du clustering est de diviser les observations en groupes de manière à ce que les observations d'un même groupe soient similaires entre elles, tandis que les observations de groupes différents soient le plus dissimilaires possible. Par exemple, un modèle de clustering peut être utilisé pour regrouper des clients d'une entreprise en fonction de leurs caractéristiques (âge, revenu, etc.) afin de mieux cibler les campagnes marketing.

#### Question 1.5: Définissez ce qu'est un arbre de décision et décrivez  de manière simple sa construction.

Réponse :
Un arbre de décision est un modèle de machine learning utilisé pour effectuer des prédictions en utilisant un arbre de décision binaire. L'arbre de décision est construit de manière récursive en effectuant un processus de "découpage" de la base de données en sous-ensembles de plus en plus petits en se basant sur des tests de condition sur les caractéristiques des observations.

Voici comment un arbre de décision est construit de manière simple :

- Sélectionner une caractéristique à tester.
- Déterminer un seuil de coupure pour cette caractéristique.
- Diviser la base de données en deux sous-ensembles en fonction de si la caractéristique est supérieure ou inférieure au seuil de coupure.
- Répéter les étapes 1 à 3 pour chaque sous-ensemble jusqu'à ce que les sous-ensembles soient "purs" (c'est-à-dire qu'ils ne contiennent qu'une seule classe).

Chaque "nœud" de l'arbre de décision correspond à un test de condition sur une caractéristique, tandis que les "feuilles" de l'arbre correspondent aux prédictions finales. Lorsqu'un arbre de décision est utilisé pour faire une prédiction, l'observation est "traversée" de haut en bas de l'arbre en suivant les tests de condition jusqu'à atteindre une feuille, qui fournit la prédiction finale.


#### Question 1.6: Dans quelle situation parle-t-on "d'overfitting"?

Réponse :

On parle d'"overfitting" (ou surapprentissage) lorsqu'un modèle de machine learning est trop spécifique à la base de données d'entraînement et ne généralise pas bien aux données de test ou aux nouvelles observations. En d'autres termes, le modèle "apprend" les bruits et les erreurs de la base de données d'entraînement au lieu de capturer la relation sous-jacente entre les caractéristiques et la cible.

L'overfitting est un problème courant en apprentissage automatique et peut se produire lorsque le modèle est trop complexe par rapport à la taille de la base de données d'entraînement, ou lorsque la base de données comprend trop peu d'exemples pour permettre au modèle de généraliser correctement. Pour éviter l'overfitting, il est recommandé de disposer d'une base de données d'entraînement suffisamment grande et de choisir un modèle adapté à la complexité des données.

Cela donne une idée de la façon du modèle qui sera en mesure de généraliser les prédictions à de nouvelles données. Si on utilise toutes les données pour entraîner le modèle, il y a un risque que le modèle surajuste (overfit) aux données d'entraînement, ce qui peut entraîner de mauvaises performances sur de nouvelles données. En séparant les données en un ensemble d'entraînement et un ensemble de test, on peut évaluer la performance du modèle sur les données jamais vues auparavant et assurer que cela généralise bien les prédictions à de nouvelles données.

#### Question 2.4: Quelle métrique utilisez vous? Pourquoi? 

L'AUC est un métrique couramment utilisée pour évaluer les modèles de classification binaire. Elle mesure la capacité du modèle à distinguer entre les deux classes. Plus l'AUC est proche de 1, meilleure est la performance du modèle. L'AUC est particulièrement utile lorsque vous avez des classes déséquilibrées, c'est-à-dire lorsque l'une des classes est beaucoup plus fréquente que l'autre. Dans ce cas, il est facile pour le modèle de simplement prédire toujours la classe majoritaire et d'obtenir une précision élevée, mais cela ne reflète pas nécessairement la capacité du modèle à détecter la classe minoritaire. L'AUC permet de tenir compte de cet aspect en évaluant la performance du modèle sur chaque classe de manière indépendante.
